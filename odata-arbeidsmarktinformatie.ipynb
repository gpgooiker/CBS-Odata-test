```python
# This query takes more than a minute
WerkgelegenheidEnArbeidsduur = cbsodata.get_data('81434ned')
WerkgelegenheidEnArbeidsduur[0:4]
```

```python
WerkgelegenheidEnArbeidsduurPerioden = []

for val in WerkgelegenheidEnArbeidsduur:
    WerkgelegenheidEnArbeidsduurPerioden.append(val['Perioden'])
    
# '2025' seems to be the latest. There's a lot of special strings in the Perioden field, for
# example '2010 na methodewijziging'
WerkgelegenheidEnArbeidsduurPerioden
```

```python
WerkgelegenheidEnArbeidsduurFiltered = [
    val for val in WerkgelegenheidEnArbeidsduur if 
        val['Perioden'] == '2025' and
        val['BedrijfstakkenBranchesSBI2008'] == 'B Delfstoffenwinning' and
        val['GeslachtWerknemer'] == 'Totaal mannen en vrouwen' and
        val['Dienstverband'] == 'Totaal' and
        val['OverigeKenmerkenWerknemer'] == 'Totaal'
]

# Sector size for 'B Delfstoffenwinning' in thousands in 2025
WerkgelegenheidEnArbeidsduurFiltered[0]['Banen_1']
```
